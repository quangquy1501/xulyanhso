```python
import cv2
import numpy as np
import random
from matplotlib import pyplot as plt
import os
def inverse_transform(img):
    return 255 - img

def gamma_correction(img):
    gamma = random.uniform(0.5, 2.0)
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    print(f"Gamma value used: {gamma:.2f}")
    return cv2.LUT(img, table)

def log_transform(img):
    c = random.uniform(1.0, 5.0)
    img_float = img.astype(np.float32)
    log_img = c * np.log1p(img_float)
    log_img = np.clip(log_img, 0, 255)
    print(f"Log coefficient used: {c:.2f}")
    return log_img.astype(np.uint8)

def hist_equalization(img):
    if len(img.shape) == 2:
        return cv2.equalizeHist(img)
    else:
        img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
        img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
        return cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

folder_path = "New folder"
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

print("Danh sách ảnh trong thư mục 'New folder':")
for idx, fname in enumerate(image_files):
    print(f"{idx+1}. {fname}")

choice_img = int(input("Nhập số thứ tự ảnh muốn chọn: ")) - 1
img_path = image_files[choice_img]
img = cv2.imread(img_path)
print(f"Đã chọn ảnh: {img_path}")

print("Chọn phương pháp biến đổi:")
print("1. Image inverse transformation")
print("2. Gamma-Correction")
print("3. Log Transformation")
print("4. Histogram equalization")
choice = int(input("Nhập số thứ tự phương pháp: "))

if choice == 1:
    result = inverse_transform(img)
    title = "Inverse"
elif choice == 2:
    result = gamma_correction(img)
    title = "Gamma Correction"
elif choice == 3:
    result = log_transform(img)
    title = "Log Transformation"
elif choice == 4:
    result = hist_equalization(img)
    title = "Histogram Equalization"
else:
    print("Lựa chọn không hợp lệ.")
    result = None

if result is not None:
    plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```
