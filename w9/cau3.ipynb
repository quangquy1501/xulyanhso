```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('colorful-ripe-tropical-fruits.jpg')
h, w = img.shape[:2]
new_w = w + 30
new_h = h + 30
resized_img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
cv2.imwrite('output_resize_fruits.jpg', resized_img)
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
img = cv2.imread('quang_ninh.jpg')
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
flipped = cv2.flip(rotated, 1)
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
cv2.imwrite('output_quangninh_rotated_flipped.jpg', flipped)
```

```python
import cv2
import numpy as np

image1 = cv2.imread('colorful-ripe-tropical-fruits.jpg')
height1, width1 = image1.shape[:2]
resized_image1 = cv2.resize(image1, (width1 + 30, height1 + 30))
cv2.imwrite('output_fruits_resized.jpg', resized_image1)

image2 = cv2.imread('quang_ninh.jpg')
height2, width2 = image2.shape[:2]
center = (width2 // 2, height2 // 2)
rotate_matrix = cv2.getRotationMatrix2D(center, -45, 1.0)
rotate_matrix[0, 2] += 50
rotated_image2 = cv2.warpAffine(image2, rotate_matrix, (width2, height2))
cv2.imwrite('output_quangninh_rotated.jpg', rotated_image2)

image3 = cv2.imread('pagoda.jpg')
resized_image3 = cv2.resize(image3, None, fx=5, fy=5, interpolation=cv2.INTER_LINEAR)
blurred_image3 = cv2.GaussianBlur(resized_image3, (7, 7), 0)
cv2.imwrite('output_pagoda_blurred.jpg', blurred_image3)
alpha = 1.5  
beta = 30    
adjusted_image3 = cv2.convertScaleAbs(blurred_image3, alpha=alpha, beta=beta)
cv2.imwrite('output_pagoda_adjusted.jpg', adjusted_image3)
```
