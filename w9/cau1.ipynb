```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('a.jpg')
mean_filtered = cv2.blur(img, (3, 3))
plt.imshow(cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB))
plt.title('Mean Filtered')
plt.axis('off')
plt.show()
```

```python
from matplotlib import pyplot as plt

img = cv2.imread('a.jpg', 0)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobelx, sobely)
plt.imshow(sobel, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import random

img = cv2.imread('a.jpg')
channels = [0, 1, 2]
random.shuffle(channels)
random_color_img = img[:, :, channels]
cv2.imwrite('a_random_color.jpg', random_color_img)
```

```python
import cv2
img = cv2.imread('a.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

cv2.imwrite('a_hue.jpg', h)
cv2.imwrite('a_saturation.jpg', s)
cv2.imwrite('a_value.jpg', v)
```
