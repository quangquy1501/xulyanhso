```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Đọc ảnh mức xám
a = cv2.imread('balloons_noisy.png', cv2.IMREAD_GRAYSCALE)

# Khử nhiễu bằng median filter trước khi xác định biên
a_denoised = cv2.medianBlur(a, 5)

# Áp dụng Sobel để xác định biên
sobelx = cv2.Sobel(a_denoised, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(a_denoised, cv2.CV_64F, 0, 1, ksize=3)
b = cv2.magnitude(sobelx, sobely).astype(np.uint8)

cv2.imwrite('sobel_balloons_noisy.png', b)
plt.imshow(b, cmap='gray')
plt.axis('off')
plt.show()
```
