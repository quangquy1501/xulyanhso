```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
img = cv2.imread('baby.jpeg')  
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# p dng mean filter (blur)
mean_filtered = cv2.blur(img, (5, 5))  # kernel 5x5

# Hin th nh gc v nh  lc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('nh gc')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Mean Filter')
plt.imshow(mean_filtered)


plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# p dng median filter (lc trung v)
median_filtered = cv2.medianBlur(img, 5)  # kernel size 5

# Hin th nh gc v nh  lc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('nh gc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Median Filter')
plt.imshow(median_filtered)
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh gc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hm p dng max filter (lc cc i)
def max_filter(img, ksize):
    return cv2.dilate(img, np.ones((ksize, ksize), np.uint8))

# p dng max filter vi kernel size 5
max_filtered = max_filter(img, 5)

# Hin th nh gc v nh  lc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('nh gc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Max Filter')
plt.imshow(max_filtered)
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh gc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hm p dng min filter (lc cc tiu)
def min_filter(img, ksize):
    return cv2.erode(img, np.ones((ksize, ksize), np.uint8))

# p dng min filter vi kernel size 5
min_filtered = min_filter(img, 5)

# Hin th nh gc v nh  lc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('nh gc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Min Filter')
plt.imshow(min_filtered)
plt.axis('off')

plt.show()
```
