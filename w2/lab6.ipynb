```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
img = cv2.imread('baby.jpeg')  
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Áp dụng mean filter (blur)
mean_filtered = cv2.blur(img, (5, 5))  # kernel 5x5

# Hiển thị ảnh gốc và ảnh đã lọc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Ảnh gốc')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Mean Filter')
plt.imshow(mean_filtered)


plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Áp dụng median filter (lọc trung vị)
median_filtered = cv2.medianBlur(img, 5)  # kernel size 5

# Hiển thị ảnh gốc và ảnh đã lọc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Ảnh gốc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Median Filter')
plt.imshow(median_filtered)
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh gốc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hàm áp dụng max filter (lọc cực đại)
def max_filter(img, ksize):
    return cv2.dilate(img, np.ones((ksize, ksize), np.uint8))

# Áp dụng max filter với kernel size 5
max_filtered = max_filter(img, 5)

# Hiển thị ảnh gốc và ảnh đã lọc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Ảnh gốc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Max Filter')
plt.imshow(max_filtered)
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh gốc
img = cv2.imread('baby.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hàm áp dụng min filter (lọc cực tiểu)
def min_filter(img, ksize):
    return cv2.erode(img, np.ones((ksize, ksize), np.uint8))

# Áp dụng min filter với kernel size 5
min_filtered = min_filter(img, 5)

# Hiển thị ảnh gốc và ảnh đã lọc
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Ảnh gốc')
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Min Filter')
plt.imshow(min_filtered)
plt.axis('off')

plt.show()
```
