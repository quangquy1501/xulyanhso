```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('bird.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
B, G, R = cv2.split(image)
red_image = cv2.merge([np.zeros_like(B), np.zeros_like(G), R])
green_image = cv2.merge([np.zeros_like(B), G, np.zeros_like(R)])
blue_image = cv2.merge([B, np.zeros_like(G), np.zeros_like(R)])
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(red_image, cv2.COLOR_BGR2RGB))
plt.title('Anh Đỏ')
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(green_image, cv2.COLOR_BGR2RGB))
plt.title('Anh Xanh Lá')
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB))
plt.title('Anh Xanh Dương')
plt.axis("off")
plt.tight_layout()
plt.show()
cv2.imwrite('anh_do.jpg', red_image)
cv2.imwrite('anh_xanh_la.jpg', green_image)
cv2.imwrite('anh_xanh_duong.jpg', blue_image)
print("Red, Green, and Blue channel images saved as 'anh_do.jpg', 'anh_xanh_la.jpg', and 'anh_xanh_duong.jpg' respectively.")
```
